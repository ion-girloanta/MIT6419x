```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
```

## L4-Correlation

```python
Xs = np.array([0.0339, 0.0423, 0.213, 0.257, 0.273, 0.273, 0.450, 0.503, 0.503, \
0.637, 0.805, 0.904, 0.904, 0.910, 0.910, 1.02, 1.11, 1.11, 1.41, \
1.72, 2.03, 2.02, 2.02, 2.02])

Ys = np.array([-19.3, 30.4, 38.7, 5.52, -33.1, -77.3, 398.0, 406.0, 436.0, 320.0, 373.0, \
93.9, 210.0, 423.0, 594.0, 829.0, 718.0, 561.0, 608.0, 1.04E3, 1.10E3, \
840.0, 801.0, 519.0])

N = 24
sns.scatterplot(x=Xs,y=Ys)
plt.show()
```

```python
print('X_mean=',np.mean(Xs), np.sum(Xs)/N)
print('Y_mean=',np.mean(Ys), np.sum(Ys)/N)
print('cov=',np.cov(Xs,Ys), 1/(N-1)*np.sum((Xs-np.mean(Xs))*(Ys-np.mean(Ys))))
print('correlation coefficient=',np.corrcoef(Xs,Ys),np.sum((Xs-np.mean(Xs))*(Ys-np.mean(Ys)))/23/(np.std(Xs,ddof=1)*np.std(Ys,ddof=1)) )
```

## L4-correlation coefficient

```python
x_distribution=[np.random.random(200), np.arange(-100,100), 
                np.linspace(norm.ppf(0.01),norm.ppf(0.99), 200),
               norm.rvs(size=200),np.random.randn(200)]
for x in x_distribution:
    y=x**2
    print(np.corrcoef(x,y))
    sns.histplot(x=x)
    plt.show()
    sns.scatterplot(x=x,y=y)
    plt.show()
```

## L4-Regression

```python
beta_1_H = (np.sum((Xs-np.mean(Xs))*(Ys-np.mean(Ys))))\
            /np.sum((Xs-np.mean(Xs))**2)
beta_0_H = (np.mean(Ys))-beta_1_H*np.mean(Xs)
print(beta_1_H,beta_0_H)
```

### Goodness of fit

```python
def y_pred(X):
    return beta_1_H*X+beta_0_H
y_p= y_pred(Xs)
1-np.sum((Ys-y_p)**2)/np.sum((Ys-np.mean(Ys))**2)
```
